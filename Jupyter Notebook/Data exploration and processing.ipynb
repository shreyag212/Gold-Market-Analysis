## Other Sheets

```python
def get_data(Sheet_Name):
    excel = pd.read_excel('prices.xlsx', sheet_name= Sheet_Name)
    
    currencies = excel.iloc[5,3:].values
    currencies[0] = 'Date'
    
    excel = excel.iloc[8:,3:]
    excel.columns = currencies
    
    excel = excel.loc[:,~excel.columns.duplicated()]
    
    return excel
```

get data from sheets

```python
annual_eop = get_data('Annual_EndofPeriod')
quarterly_eop = get_data('Quarterly_EndofPeriod')
monthly_eop = get_data('Monthly_EndofPeriod')
weekly_eop = get_data('Weekly_EndofPeriod')
daily = get_data('Daily')
daily_index = get_data('Daily_Indexed')
```

## Checking similar dataframes (Monthly Average vs Monthly EndofPeriod, Daily vs Daily Index)

```python
print(monthly_eop.columns.values)
cols = monthly_eop.columns.values.tolist()
monthly_avg.columns = cols
```

```python
fig, ax = plt.subplots()
for cur in cols[1:]:
    monthly_eop.plot(x='Date', y=cur, ax=ax, legend = False, color = 'C0')
    monthly_avg.plot(x='Date', y=cur, ax=ax, legend = False, color = 'C1')
plt.ylim([0,2000])
```

```python
for cur in cols[1:]:
    percentage_diff = abs((monthly_avg[cur] - monthly_eop[cur])/monthly_avg[cur]*100)
    percentage_diff.plot()
plt.ylim([0,100])
```

```python
percentage_diff.plot()
plt.ylim([0,100])
```

```python
fig, ax = plt.subplots()
for cur in cols[1:]:
    daily.plot(x='Date', y=cur, ax=ax, legend = False, color = 'C0')
    daily_index.plot(x='Date', y=cur, ax=ax, legend = False, color = 'C1')
```

# time stamp conversion

```python
daily['Date'] = pd.to_datetime(daily['Date'])
daily['day'] = daily['Date'].apply(lambda time: time.dayofweek)
daily['week'] = daily['Date'].apply(lambda time: time.week)
daily['month'] = daily['Date'].apply(lambda time: time.month)
daily['Quarter'] = daily['Date'].apply(lambda time: int(time.month/12 +1))
daily['year'] = daily['Date'].apply(lambda time: time.year)
```

```python
daily.head(4)
```

```python
weekly_eop['Date'] = pd.to_datetime(weekly_eop['Date'])
weekly_eop['week'] = weekly_eop['Date'].apply(lambda time: time.week)
weekly_eop['month'] = weekly_eop['Date'].apply(lambda time: time.month)
weekly_eop['Quarter'] = weekly_eop['Date'].apply(lambda time: int(time.month/12 +1))
weekly_eop['year'] = weekly_eop['Date'].apply(lambda time: time.year)
weekly_eop.head(4)
```

```python
monthly_eop['Date'] = pd.to_datetime(monthly_eop['Date'])
monthly_eop['month'] = monthly_eop['Date'].apply(lambda time: time.month)
monthly_eop['Quarter'] = monthly_eop['Date'].apply(lambda time: int(time.month/12 +1))
monthly_eop['year'] = monthly_eop['Date'].apply(lambda time: time.year)
monthly_eop.head(4)
```

```python
quarterly_eop['Date'].iloc[0]
```

```python
quarterly_eop['Quarter'] = quarterly_eop['Date'].apply(lambda time: time[1])
quarterly_eop['year'] = quarterly_eop['Date'].apply(lambda time: time[3:])
quarterly_eop.head(4)
```
