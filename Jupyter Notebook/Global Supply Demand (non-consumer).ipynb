```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Data Source: <a href = https://www.gold.org/goldhub/data/gold-supply-and-demand-statistics> Goldhub

```python
sheet = pd.read_excel('gdt-q2-2021-statistics_EN (1).xlsx', sheet_name = 'Gold Balance')
sheet.head(20)
```

```python
sup_dem = sheet.loc[[3,8,18,19]]
sup_dem
```

```python
sup_dem = sup_dem.iloc[:,16:-2]
sup_dem.rename(index = {3:'Term', 8:'Supply', 18:'Demand', 19:'left'}, inplace = True)
sup_dem = sup_dem.T.reset_index(drop = True)
sup_dem.head(9)
```

```python
sup_dem['Supply'] = sup_dem['Supply'].apply(lambda x: round(x,2))
sup_dem['Demand'] = sup_dem['Demand'].apply(lambda x: round(x,2))
sup_dem['left'] = sup_dem['left'].apply(lambda x: round(x,2))
sup_dem.head(9)
```

```python
def quarter_format(qtr):
    qtr = '20' + qtr[3:] + '-' + qtr[:2]
    return qtr

def term_to_date(df):
    df['Term'] = df['Term'].apply(quarter_format)
    df['Term'] = pd.PeriodIndex(df['Term'], freq='Q').to_timestamp()
    df['Term'] = pd.to_datetime(df['Term']).dt.date
    return df

quarter_format("Q1'10")
```

```python
sup_dem = term_to_date(sup_dem)
sup_dem.head(9)
```

```python
fig,ax = plt.subplots(figsize = (12,6))

sup_dem.plot(x = 'Term', y = 'Demand', ax=ax, label='Demand')
sup_dem.plot(x = 'Term', y = 'Supply', ax=ax, label='Supply')
sup_dem.plot(x = 'Term', y = 'left', ax=ax, label='Surplus')

plt.title('Supply Demand over Time')
plt.ylabel('Weigth (tonnes)')
plt.xlabel('Time (Quarterly)')
plt.show()
```
