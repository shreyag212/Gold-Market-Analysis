## Other Sheets

```python
def get_data(Sheet_Name):
    sheet = pd.read_excel('prices.xlsx', sheet_name= Sheet_Name)
    
    prices = sheet.iloc[8:,3:]
    prices.columns = sheet.iloc[5,3:].values
    prices = prices.loc[:,~prices.columns.duplicated()]
    prices.rename(columns = {np.nan: 'Date', 'Name': 'Date'}, inplace = True)
    return prices
```

get data from sheets

```python
annual = get_data('Annual_EndofPeriod')
quarterly = get_data('Quarterly_EndofPeriod')
monthly_eop = get_data('Monthly_EndofPeriod')
weekly = get_data('Weekly_EndofPeriod')
daily = get_data('Daily')
daily_index = get_data('Daily_Indexed')
```

## Checking similar dataframes (Monthly Average vs Monthly EndofPeriod, Daily vs Daily Index)

```python
cols = monthly_eop.columns.values.tolist()
monthly_avg.columns = cols

for cur in cols[1:]:
    percentage_diff = abs((monthly_avg[cur] - monthly_eop[cur])/monthly_avg[cur]*100)
    percentage_diff.plot()
    
plt.title('% Error in monthly Avg vs EOP over Index')
plt.xlabel('Index')
plt.ylabel('% Error')
plt.ylim([0,100])
plt.show()
```

### Daily vs Daily Indexed data

```python
fig, ax = plt.subplots()
for cur in cols[1:]:
    daily.plot(x='Date', y=cur, ax=ax, legend = False, label = 'Normal', color = 'C0')
    daily_index.plot(x='Date', y=cur, ax=ax, legend = False, label = 'Indexed', color = 'C1')
    if cur == cols[1]:
        plt.legend()

plt.title('Daily Prices (normal and indexed)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
import statistics
```

```python
mean_index = []
stdev_index = []
for cur in cols[1:]:
    ratio = daily[cur].dropna()/daily_index[cur].dropna()
    mean_index.append(statistics.mean(ratio))
    stdev_index.append(statistics.stdev(ratio))
    ratio.plot()
    
plt.ylim([0,10])
plt.title('Daily (Price/Indexed Price)')
plt.xlabel('Data Points')
plt.ylabel('Ratio')
plt.show()
```

```python
currency_index = pd.DataFrame([mean_index, stdev_index], columns = cols[1:], index = ['mean', 'stdev']).T
print(currency_index)
```

# time stamp conversion

```python
daily['Date'] = pd.to_datetime(daily['Date']).dt.date
weekly['Date'] = pd.to_datetime(weekly['Date'])
monthly_eop['Date'] = pd.to_datetime(monthly_eop['Date'])
monthly_avg['Date'] = pd.to_datetime(monthly_avg['Date'])
```

```python
daily.head(3)
```

```python
monthly_eop.head(3)
```

```python
def quarter_format(qtr):
    qtr = qtr[3:] + '-' + qtr[:2]
    return qtr


quarterly['Date'] = quarterly['Date'].apply(quarter_format)
quarterly['Date'] = pd.PeriodIndex(quarterly['Date'], freq='Q').to_timestamp()

quarterly.head(4)
```

I suspect there is missing data for many currencies before 01/02/1979

```python
daily.drop([8,9], inplace = True)
weekly.drop(8,inplace = True)
monthly_eop.drop(8,inplace = True)
monthly_avg.drop(8,inplace = True)
quarterly.drop(8,inplace = True)
annual.drop(8,inplace = True)
```

```python
arr = np.zeros(6)
arr = arr.reshape(2,3)
quarter = pd.DataFrame(arr, columns = ['Date', 'Currency','Price'])

for i in range(quarterly.shape[0]):
    for j in range(1,quarterly.shape[1]):
        date = quarterly.iloc[i,0]
        currency = quarterly.columns.values[j]
        price = quarterly.iloc[i,j]
        temp = {'Date': date, 'Currency': currency, 'Price': price}
        quarter = quarter.append(temp, ignore_index=True)

quarter.drop([0,1], inplace=True)
quarter.reset_index(inplace=True, drop=True)    
quarter['Date'] = pd.to_datetime(quarter['Date']).dt.date
quarter    
```

```python
fig,ax = plt.subplots()
for i in quarter['Currency'].unique():
    quarter[quarter['Currency'] == i].plot(x='Date', y='Price', ax=ax, legend = False)
    
plt.ylim([0,10000])
plt.title('Gold Price Trends for Major Currencies')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
